## Imports ##

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("openpowerlifting-2025-09-27-31932eca.csv")

print(df.columns)
```

```python
cols = ['Sex', 'Age', 'BodyweightKg', 'WeightClassKg', 'Federation', 'Equipment', ]

# Event: to only SBD
# Equipment: Raw + Wraps, 
# Age: drop empty fields
# TotalKg: drop empty fields
# Place: must be a number (no DQs)
# parent federation to only IPF

data = df[
    (df['Event'] == 'SBD') & 
    (df['Equipment'].isin(['Raw', 'Wraps'])) &
    (df['Age'].notna()) &
    (df['Division'] == 'Open') &
    (df['TotalKg'].notna()) &
    (df['Place'].str.isnumeric()) &
    (df['ParentFederation'] == 'IPF')
    # (df['MeetCountry'].isin(['Scotland', 'England', 'USA', ]))
    ]

data.to_csv("cleanIPF.csv", index=False)
```
